```python
results=[]
trT0=[]
trTLH=[]
for L in Ls:
    trT = np.loadtxt("obs_L"+str(L)+"_0/eps_jk.dat")
    results.append([L,trT])
    trT0.append(trT[-1])
    trTLH.append(trT[int(L/2)])
```

```python
def linear(x, a, b):
    return a*x + b
```

```python
fitm=4
# fitM=

opt = sp.optimize.curve_fit( linear, ells[fitm:], np.array(trT0).T[0][fitm:] )
```
