```python
crr="ss"
crr="sseps"
# crr="ssT11"
```

```python
results=[]
for L in Ls:
    corr = np.loadtxt("obs_L"+str(L)+"_0/"+crr+"_jk.dat")
    results.append([L,corr])
```

```python
markers=("x","+","*",".")
colors=("red","green","magenta","blue")
```

```python
i=1
```

```python
x = results[i][1].T[0]
y = results[i][1].T[1]
z = results[i][1].T[2]
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, np.log(z) )
# ax.scatter(x, y, np.log(-z) )
```

```python
for i in np.arange(len(Ls))[::-1]:
    plt.errorbar( results[i][1].T[0], results[i][1].T[1]/ells[i]**2, results[i][1].T[2]/ells[i]**2, 
                 ls='none', marker=markers[i], color=colors[i], capsize=2, label='L='+str(Ls[i]) )

# xx=np.linspace(0.0, np.pi, 200)
# yy=(2.0/(1.0-np.cos(xx)))**(1.0/8.0)
# plt.plot(xx, yy, ls='dashed', label='analytic')

plt.yscale("log")
# plt.xscale("log")
plt.legend()

plt.xlabel("$\\vert \\Omega \\vert$")
plt.ylabel("$\\langle \\sigma(O) \\sigma(\\Omega) \\rangle$")

# plt.savefig("ss_corr.pdf", bbox_inches='tight')
```
